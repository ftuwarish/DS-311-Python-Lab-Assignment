## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Create a new column to show budget per student in each row
school_data_complete["spending per student"] = school_data_complete["budget"] / school_data_complete["size"]

```

```python
# Create a new column to define the spending ranges per student
school_data_complete["spending"] = pd.cut(
    school_data_complete["spending per student"], spending_bins, labels=group_names)

```

```python
# Calculate the average math score within each spending range
bins = [0, 500, 1000, 1500, 2000, float('inf')]
labels = ['<500', '500-1000', '1000-1500', '1500-2000', '2000+']
school_data_complete['Spending Range'] = pd.cut(school_data_complete['spending per student'], bins=bins, labels=labels)


```

```python
# Calculate the percentage passing rate for math in each spending range
spending_math_pass = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
math_pass_sp = spending_math_pass.groupby(['Spending Range'], observed=False)['math_score'].count().reset_index(name='passing_math_count')
total_math_counts = school_data_complete.groupby(['Spending Range'], observed=False)['math_score'].count().reset_index(name='total_math_count')
math_pass_percent_df = pd.merge(math_pass_sp, total_math_counts, on='Spending Range', how='left')
math_pass_percent_df['percent_passing_math'] = (math_pass_percent_df['passing_math_count'] / math_pass_percent_df['total_math_count']) * 100
```

```python
# Calculate the percentage passing rate for reading in each spending range
spending_read_pass = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
read_pass_sp = spending_read_pass.groupby(['Spending Range'])['reading_score'].count().reset_index(name='passing_reading_count')
total_read_counts = school_data_complete.groupby(['Spending Range'])['reading_score'].count().reset_index(name='total_read_count')
read_pass_percent_df = pd.merge(read_pass_sp, total_read_counts, on='Spending Range', how='left')
read_pass_percent_df['percent_passing_reading'] = (read_pass_percent_df['passing_reading_count'] / read_pass_percent_df['total_read_count']) * 100
```

```python
# Calculate the percentage overall passing rate in each spending range
overall_pass_percent_df = pd.merge(math_pass_percent_df, read_pass_percent_df,
                                    on='Spending Range', how='left')

# Calculate the overall passing percentage as the average of math and reading passing percentages
overall_pass_percent_df['overall_passing_rate'] = (
    overall_pass_percent_df['percent_passing_math'] + overall_pass_percent_df['percent_passing_reading']
) / 2

# Select the relevant columns to display
result_overall_df = overall_pass_percent_df[['Spending Range', 
                                              'percent_passing_math', 
                                              'percent_passing_reading', 
                                              'overall_passing_rate']]

# Display the final DataFrame
print(result_overall_df)
```
